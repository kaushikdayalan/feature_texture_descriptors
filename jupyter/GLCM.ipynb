```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np
import pywt
```

```python
sky_image = cv2.imread("../images/sky.jpeg")

gray_sky_image = cv2.cvtColor(sky_image, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(cv2.cvtColor(sky_image, cv2.COLOR_BGR2RGB))
```

```python

# get img greyscale

# choose a positional operator
pos_op = [1,0]

# init glcm array
glcm = np.zeros_like(gray_sky_image)

# iterate over image and complete glcm
for i in range(gray_sky_image.shape[0]): # row
    for j in range(gray_sky_image.shape[1]): # col
        init_val = gray_sky_image[i,j]
        try:
            target = gray_sky_image[i+pos_op[0],j+pos_op[1]]
        except IndexError:
            continue # out of img bounds
        glcm[init_val,target]+=1

glcm = glcm/np.sum(glcm)

plt.imshow(np.log(glcm+1e-6), cmap="gray")
plt.show()
```
